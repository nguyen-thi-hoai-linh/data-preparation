```python
#Ex3: Calculate the mean of mag in each quarter

from datetime import datetime
data['time_2'] = data['time']
data['time_2'] = pd.to_datetime(data['time_2'],unit = 'ms')
data['month'] = pd.DatetimeIndex(data['time_2']).month
q1 = data[data['month']<=3]
print(q1['mag'].mean())
q2 = data[(data['month']>=4)& (data['month']<=6)]
print(q2['mag'].mean())
q3 = data[(data['month']>=7)& (data['month']<=9)]
print(q3['mag'].mean())
q4 = data[(data['month']>=10)& (data['month']<=12)]
print(q4['mag'].mean())
```
